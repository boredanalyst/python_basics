# Data Structures

```python
import pandas as pd
import numpy as np
```

## Lists

```python
## Lists are ordered data structures.
## Declare lists with square brackets

mylist = ["apple","orange","banana"]
mylist
```

```python
## Lists are ordered so you can access items using indexing.

mylist[0] ## This returns the first item in the list.
```

```python
## Lists are mutable so you can change the items in the list.

mylist[0] = "orange" ## This changes the first item of the list to orange

mylist
```

## Tuples

```python
## Tuples are immutable collection of items.
## Declare tuples with parentheses.

mytuple = ("apple","orange","banana")

mytuple
```

```python
## You cannot change the elements of a tuple.

##mytuple[0] = "mango"
```

## Sets

```python
## Sets are collection of unique data items.
## You can declare sets with curly braces.

myset = {"apple","orange","banana"}

myset
```

```python
## Sets do not allow duplicates.

myset = {"apple","orange","banana","apple"}

myset
```

## Dictionaries

```python
## Dictionaries are collections of key-value pairs.
## You can declare dictionaries using curly braces and colons.

mydict = {"Fruit":"Apple","Price":120}

mydict
```

```python
## You can use the keys to access the elements of a dictionary.

mydict["Fruit"]
```
